# Question 1- Report Card:

```python
marks = float(input("Enter Student marks: \n"))
print("\nStudent grade is:")
if marks>=90:
    print("A+")
elif marks>=80 and marks<90:
    print("A")
elif marks>=70 and marks<80:
    print("B+")
elif marks>=60 and marks<70:
    print("B")    
elif marks>=50 and marks<60:
    print("C+")    
elif marks>=40 and marks<50:
    print("C")
else: 
    print("F")    
```

# Question 2- Print Prime Numbers between 1 and 1000:

```python
flag = 0 
print("The Prime Numbers between 1 and 1000 are: \n")
for item in range(1,1000):        #generating numbers from 1 to 999
    for item2 in range(1,1000):   #generating numbers from 1 to 999 again for division
        if(item%item2 == 0):      #if the numbers are divisible and return 0 as remainder:
            flag += 1             # we increment the flag state by 1
    if(flag == 2):                # if flag = 2 i.e divisible only by 1 and itself
        print(item)               # we print :)
    flag = 0                      # we are updating the value of flag to zero after every iteration so that our logic works perfectly!
        
        
```

# Question 3- Program to print Number tables from 1 to 10:

```python
for multiple in range(1,11):       # Generating numbers from 1 to 10
    print(f"The Multiplication table for {multiple} is:\n")   #Using f-string to print Header 
    for multiplier in range(1,11):                   #Generating numbers from 1 to 10
        print(f" {multiple}x{multiplier}= {multiple*multiplier}\n")   #using f-string to print result in nested loop
        
```
